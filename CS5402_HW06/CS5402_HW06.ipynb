```python
from surprise import Dataset
from surprise import Reader
from surprise import SVD
from surprise import evaluate, print_perf
import os

file_path = os.path.expanduser('restaurant_ratings.txt')
reader = Reader(line_format='user item rating timestamp', sep='\t')
data = Dataset.load_from_file(file_path, reader=reader)
data.split(n_folds=3)
results = []
```

```python
algo = SVD()
perf = evaluate(algo, data, measures=['RMSE', 'MAE'])
print_perf(perf)
results.append(('SVD',perf))
```

```python
algo = SVD(biased=False)
perf = evaluate(algo, data, measures=['RMSE', 'MAE'])
print_perf(perf)
results.append(('PMF',perf))
```

```python
from surprise import NMF
algo = NMF()
perf = evaluate(algo, data, measures=['RMSE', 'MAE'])
print_perf(perf)
results.append(('NMF',perf))
```

```python
from surprise import KNNBasic
algo = KNNBasic(sim_options = {'user_based':True})
perf = evaluate(algo, data, measures=['RMSE', 'MAE'])
print_perf(perf)
results.append(('User-based collaborative filtering',perf))
```

```python
algo = KNNBasic(sim_options = {'user_based':False})
perf = evaluate(algo, data, measures=['RMSE', 'MAE'])
print_perf(perf)
print(perf)
results.append(('Item-based collaborative filtering',perf))
```

```python
for comparator in ('MSD', 'cosine', 'pearson'):
    algo = KNNBasic(sim_options = {'user_based':True, 'name': comparator})
    perf = evaluate(algo, data, measures=['RMSE', 'MAE'])
    print_perf(perf)
    results.append(('User-based collaborative filtering with {0}'.format(comparator),perf))
```

```python
import pandas as pd
final_results = {'model_name': []}

for model_name, perf in results:
    final_results['model_name'].append(model_name)
    
    for metric in perf.keys():
        for i, score in enumerate(perf[metric]):
            fold_name = 'fold{0}_{1}'.format(i+1, metric)
            if fold_name not in final_results:
                final_results[fold_name] = []
            
            final_results[fold_name].append(score)
print(final_results)
results_df = pd.DataFrame.from_dict(final_results)
```

```python

results_df.reindex(sorted(results_df.columns), axis=1)
results_df.head()
```

```python
results_df['mean_rmse'] = results_df[['fold{0}_rmse'.format(i) for i in range(1, 4)]].mean(axis=1)
results_df['mean_mae'] = results_df[['fold{0}_mae'.format(i) for i in range(1, 4)]].mean(axis=1)
results_df[:]
```

```python

def get_top(df, col_name):
    return df['model_name'].iloc[df[col_name].idxmin()]

def summarize_fold(df, idx):
    metrics = {'rmse': 'fold{0}_rmse'.format(idx), 'mae': 'fold{0}_mae'.format(idx),
               'mean rmse': 'mean_rmse', 'mean mae': 'mean_mae'}
    print('Summary for fold {0}:'.format(idx))
    for metric_name, key_value in metrics.items():
        print('  Best model over {0} score: {1}'.format(metric_name.upper(), get_top(df, key_value)))
        
for i in range(1, 4):
    summarize_fold(results_df, i)
```

```python
mean_rmse = []
import numpy as np

for k in range(2,50):
    algo = KNNBasic(k=k, sim_options={'name':'MSD', 'user_based':True})
    mean_rmse.append(np.asarray(evaluate(algo, data, measures=['RMSE', 'MAE'])['rmse']).mean())
```

```python
import seaborn as sns
import matplotlib 
%matplotlib inline
series = pd.Series(name='rmse', data=mean_rmse)
ax = sns.barplot(series.index, series.values)
ax.set_ylabel('RMSE')
ax.set_xlabel('k')
ax.set_xlim(0, 30)
ax.set_ylim(.9, 1.2)
```
